# Import libareires

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load the Dataset

```python
# Load the planets dataset
planets = sns.load_dataset("planets")
planets
```

```python
# Show the first few rows of the dataset
print("Dataset Overview:")
print(planets.head())
```

```python
# Get basic information about the dataset
print(planets.info())
```

```python
# Get descriptive statistics for numerical columns
print(planets.describe())
```

```python
# Check for missing values
print(planets.isnull().sum())
```

```python

# Drop rows with missing values
planets_cleaned = planets.dropna()
```

```python
# Verify that missing values are removed
print(planets_cleaned.isnull().sum())
```

```python
# Plot the distribution of planet mass
plt.figure(figsize=(8, 5))
sns.histplot(planets_cleaned['mass'], bins=20, kde=True)
plt.title("Distribution of Planet Mass")
plt.xlabel("Mass (Jupiter Mass)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot for planet mass vs orbital period
plt.figure(figsize=(8, 5))
sns.scatterplot(data=planets_cleaned, x="mass", y="orbital_period", hue="method", palette="coolwarm")
plt.title("Planet Mass vs Orbital Period")
plt.xlabel("Mass (Jupiter Mass)")
plt.ylabel("Orbital Period (days)")
plt.show()
```

```python
# Box plot for mass by detection method
plt.figure(figsize=(12, 6))
sns.boxplot(data=planets_cleaned, x="method", y="mass")
plt.title("Planet Mass Distribution by Detection Method")
plt.xticks(rotation=90)
plt.xlabel("Detection Method")
plt.ylabel("Planet Mass (Jupiter Mass)")
plt.show()
```

```python
# Count plot for detection methods
plt.figure(figsize=(12, 6))
sns.countplot(data=planets_cleaned, x="method")
plt.title("Number of Planets Detected by Method")
plt.xticks(rotation=90)
plt.xlabel("Detection Method")
plt.ylabel("Number of Planets")
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(planets_cleaned.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Planets Dataset")
plt.show()
```

```python
# Pairplot for planets dataset
sns.pairplot(planets_cleaned)
plt.show()
```

## Conclusion

```python
print("\nConclusions:")
print("- Planets detected by different methods show varied distributions of mass.")
print("- There is no strong correlation between planet mass and orbital period.")
print("- The number of planets detected varies significantly across methods, with some methods being more prolific than others.")
```
